## Finding 2: America does not seem to have much more firearm homicides than other developed countries.

### Comparison 
The original graph 1 of Vox demonstrates that America has higher firearm homicides than other developed countries.

```python
Image(filename='Graph_1.jpg', width=500)
```

The deceptive version makes America seem not to have much more firearm homicides than other developed countries.

```python
Image(filename='Decpt_Graph_1.png', width=800)
```

### Development Process
To get a deceptive graph 1 and use the new graph to prove the second finding, I intentionally picked 12 developed countries that have higher firearm homicide rate among all of the 36 developed countries categorized based on the Country Classification Report by the UN (see below source 2), so that the gap between the U.S. and other developed countries could be as smaller as possible.

Then, the x-axis is manipulated through the logarithmic scaling to make rate of the U.S. and the other developed countries look much closer. The range of the x-axis is deliberately set fixed from 0.001 to 3.5 and the largest tick mark is set as 1 to let the different countries look of similar values. 

Finally, the bar of each country is deliberately widened so that the differences in values look even smaller. 

The graph is available on Tableau Public: <br>https://public.tableau.com/profile/jieyi.deng3205#!/vizhome/Graph1Deceptive/HomicidesbyFirearm

### Reasoning
Compared with the original product, this is a deceptive version based on the following reasons:
1. The countries is intentionally chosed to show a narrower gap of firearm homicide rate between the U.S. and the other developed countries.
2. If the readers are not familiar with the concept of log scales, the graph using a log scale can be rather misleading. Such a change in scale leads to a representation that allows data values to be appered in similar size while in reality, they have widely differing magnitudes (see below source 3). 

### Data Wrangling and Data Source:

As shown above, the first step is to cherry pick the data and include countries with higher rate of firearm homicde rate to narrow down the huge gap between the U.S. and other developed countries. 

To minimize the differences visually, the log scaling is adopted to give ones the impression that there is little difference between the rate of U.S. and other developed countries selected. Specifically, the U.S. has a figure of 29.7 homicides by guns per 1 million people and the second largest one is Switzerland with a number of 7.7. The difference between these two countries is huge in the original graph, not to mention comparing the U.S. to other countries with even smaller number. On a log scale in the deceptive version, the length of these two bars are much more similar.

### Development Process

To begin with, I crated a graph with all the three locations of the original but it did not have many differences compared with the original one. 1. The y-axis is truncated with a fixed range from 44% to 61% and 31% to 50% for _Control Gun Ownership_ and _Protect Rights to Own Guns_ respectively to create an exaggeration of the trends.
